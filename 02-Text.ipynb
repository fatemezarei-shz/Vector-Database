```python
import lancedb
```

```python
uri = "c:/data/new1"
```

```python
db = lancedb.connect(uri)
```

```python
data = [
    { "vector":[1], "text": "rebel spaceships striking from a hidden base"},
    { "vector":[2],"text": "have won their first victory against the evil Galactic Empire"},
    {"vector":[3], "text": "during the battle rebel spies managed to steal secret plans"},
    { "vector":[4],"text": "to the Empire's ultimate weapon the Death Star"}
]
```

```python
tbl = db.create_table("Table3", data = data)
```

```python
tbl.create_fts_index("text", replace = True)
```

```python
results = tbl.search('text == rebel' ).limit(2).to_pandas()
```
