```python
import lancedb
```

```python
from lancedb.pydantic import LanceModel, Vector
```

```python
from lancedb.embeddings import get_registry
```

```python
db = lancedb.connect("/tmp/db")
```

```python
pip install huggingface_hub[hf_xet]
```

```python
model = get_registry().get("huggingface").create(name="BAAI/bge-m3")
```

```python
class Words(LanceModel):
    text: str = model.SourceField()
    vector: Vector(model.ndims()) = model.VectorField()

table = db.create_table("words", schema=Words)
table.add(
    [
        {"text": "نیک آموز شرکتی حرفه ای برای آموزش دیتا است."},
        {"text": "علی از بیماری سرطان رنج می برد"},
        {"text": "محمد و سحر به زودی ازدواج می کنند."}
    ]
)
```

```python
query = "تعلیم"
table.search(query).limit(1).to_pydantic(Words)
```
